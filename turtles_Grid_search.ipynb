```
#time embedding
import math
def t2v0(d):
  return math.sin((d-min).days/(max-min).days**(0/6))
def t2v1(d):
  return math.cos((d-min).days/(max-min).days**(0/6))
def t2v2(d):
  return math.sin((d-min).days/(max-min).days**(2/6))
def t2v3(d):
  return math.cos((d-min).days/(max-min).days**(2/6))
def t2v4(d):
  return math.sin((d-min).days/(max-min).days**(4/6))
def t2v5(d):
  return math.sin((d-min).days/(max-min).days**(4/6))
```

```
temp=df_fill.Date.apply(t2v0)
df_fill['DateVec0']=temp
temp=df_fill.Date.apply(t2v1)
df_fill['DateVec1']=temp
temp=df_fill.Date.apply(t2v2)
df_fill['DateVec2']=temp
temp=df_fill.Date.apply(t2v3)
df_fill['DateVec3']=temp
temp=df_fill.Date.apply(t2v4)
df_fill['DateVec4']=temp
temp=df_fill.Date.apply(t2v5)
df_fill['DateVec5']=temp
```

```
def ifweekday(d):
  return d.weekday()<=5
def weekday(d):
  return d.weekday()
```

```
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn import linear_model 
from sklearn.metrics import mean_absolute_error, mean_squared_error
import plotly

# import the relevant Keras modules
import keras
import tensorflow
from keras.models import Sequential
from keras.layers import Activation, Dense
from keras.layers import LSTM
from keras.layers import Dropout
from time import time
from sklearn.metrics import fbeta_score, make_scorer

scorer = make_scorer(mean_squared_error)

import tensorflow as tf
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import GridSearchCV
def build_model(init='glorot_uniform', optimizer='rmsprop'):
    
    model = Sequential()
    model.add(LSTM(32, input_shape=(LSTM_training_inputs.shape[1], LSTM_training_inputs.shape[2]),return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=10,kernel_initializer=init))
    model.add(Activation('linear'))
    model.compile(loss="mse", optimizer=optimizer, metrics=[tf.keras.metrics.RootMeanSquaredError()])
    model.summary()
    return model
```

```
#nn_model = build_model(LSTM_training_inputs, output_size=10, neurons = 32)
# train model on data
# note: eth_history contains information on the training error per epoch

#nn_history = nn_model.fit(LSTM_training_inputs, LSTM_training_outputs, epochs=100, batch_size=16, verbose=2, shuffle=False)
start=time()
model = KerasRegressor(build_fn=build_model)
optimizers = ['rmsprop', 'adam']
init = ['glorot_uniform', 'normal', 'uniform']
epochs = [50, 100, 150]
batches = [5, 10, 20]
param_grid = dict(optimizer=optimizers, nb_epoch=epochs, batch_size=batches, init=init)
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(LSTM_training_inputs, LSTM_training_outputs)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))

for params, mean_score, scores in grid_result.grid_scores_:
    print("%f (%f) with: %r" % (scores.mean(), scores.std(), params))
print("total time:",time()-start)
```
